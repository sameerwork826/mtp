# Domain script

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
domain_grid = np.loadtxt('domain_grid_A.dat')

# Assuming time is in the first column and subsequent columns are domain sizes
time_steps = domain_grid[:, 0]

# Plot domain size distribution at different times
for i in range(1, domain_grid.shape[1]):
    plt.plot(time_steps, domain_grid[:, i], label=f'Domain {i}')
plt.xlabel('Time')
plt.ylabel('Domain Size')
plt.title('Domain Size Distribution Over Time')
plt.legend()
plt.show()

# Calculate and plot average domain size over time
avg_domain_size = np.mean(domain_grid[:, 1:], axis=1)
plt.plot(time_steps, avg_domain_size)
plt.xlabel('Time')
plt.ylabel('Average Domain Size')
plt.title('Average Domain Size Over Time')
plt.show()
```

# Cluster script

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
cluster_dist = np.loadtxt('cluster_dist.dat')
max_cluster = np.loadtxt('max_cluster.dat')

# Assuming time is in the first column and subsequent columns are cluster sizes
time_steps = cluster_dist[:, 0]

# Plot cluster size distribution at different times
for i in range(1, cluster_dist.shape[1]):
    plt.plot(time_steps, cluster_dist[:, i], label=f'Cluster {i}')
plt.xlabel('Time')
plt.ylabel('Cluster Size')
plt.title('Cluster Size Distribution Over Time')
plt.legend()
plt.show()

# Plot largest cluster size over time
plt.plot(max_cluster[:, 0], max_cluster[:, 1])
plt.xlabel('Time')
plt.ylabel('Largest Cluster Size')
plt.title('Largest Cluster Size Over Time')
plt.show()
```
